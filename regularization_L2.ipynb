```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
import seaborn as sns
from mlxtend.plotting import plot_decision_regions

import tensorflow
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.layers import Dropout
from tensorflow.keras.optimizers import Adam
```

```python
X, y = make_moons(100, noise=0.25,random_state=2)
```

```python
sns.scatterplot(x=X[:,0],y=X[:,1],hue=y)
```

```python
model1=Sequential()
model1.add(Dense(128,input_dim=2,activation='relu'))
model1.add(Dense(128,activation='relu'))
model1.add(Dense(1,activation='sigmoid'))
model1.summary()
```

```python
adam=Adam(learning_rate=0.01)
model1.compile(loss='binary_crossentropy',optimizer=adam,metrics=['accuracy'])
```

```python
history1=model1.fit(X,y,validation_split=0.2,epochs=1000,verbose=0)
```

```python
plot_decision_regions(X,y.astype(int),clf=model1,legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,1.5)
```

```python
plt.plot(history1.history['loss'],color='red')
plt.plot(history1.history['val_loss'])
```

# Use L2 regularization

```python
model2 = Sequential()

model2.add(Dense(128,input_dim=2, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l2(0.03)))
model2.add(Dense(128, activation="relu",kernel_regularizer=tensorflow.keras.regularizers.l1(0.03)))
model2.add(Dense(1,activation='sigmoid'))

model2.summary()
```

```python
adam = Adam(learning_rate=0.01)
model2.compile(loss='binary_crossentropy', optimizer=adam, metrics=['accuracy'])

history2 = model2.fit(X, y, epochs=1000, validation_split = 0.2,verbose=0)
```

```python
plot_decision_regions(X, y.astype('int'), clf=model2, legend=2)
plt.xlim(-2,3)
plt.ylim(-1.5,2)
plt.show()
```

```python
plt.plot(history2.history['loss'])
plt.plot(history2.history['val_loss'])
```
